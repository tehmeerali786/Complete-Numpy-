# Hyperbolic Function

```python
% matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

low = -np.pi
hi = np.pi
N = 300

x = np.linspace(low, hi, N)
y = np.sinh(x)


plt.plot(x, y)
plt.show()
```

```python
y = np.cosh(x)


plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, np.sinh(x))
plt.plot(x, np.cosh(x))
plt.show()
```

```python
y = np.tanh(x)


plt.plot(x, y)
plt.show()
```

# Exponents

```python
x = np.arange(10)
y = x

plt.plot(x, y)
plt.show()
```

```python
y = np.exp(x)

plt.plot(x, y, 'o-')
plt.show()
```

```python
# exp minus 1

y = np.expm1(x)
plt.plot(x, y, 'o')
plt.show()
```

```python
# x**2

y = np.exp2(x)
plt.plot(x, y, 'o-')
plt.show()
```

```python
x = np.array([1, 3])
y = np.array([2, 4])
```
